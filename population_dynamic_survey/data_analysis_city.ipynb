```python
# !/usr/bin/env python3
# 计算一个城市的人口变化曲线
import pandas as pd
import numpy as np
from sklearn import linear_model
from sklearn import cross_decomposition
import matplotlib.pyplot as plt 
%matplotlib inline
from matplotlib.font_manager import FontProperties
def getChineseFont():  
    return FontProperties(fname='/System/Library/Fonts/PingFang.ttc',size=15)
```

```python
df_csv = pd.read_csv('./data/flow_train.csv')
df_csv['date_dt'] = df_csv['date_dt'].apply(lambda x: pd.to_datetime(str(x)))
```

```python
df_group_city = df_csv.groupby(['city_code'])
```

```python
df_group_city0 = list(df_group_city)[0][1]
```

```python
df_group_city0_dt = df_group_city0.groupby(['date_dt'])[['dwell']].sum()
```

```python
df_group_city0_dt.head()
```

```python
df_group_city0_dt = df_group_city0_dt.reset_index()
```

```python
df_group_city0_dt.head()
```

```python
fig, ax = plt.subplots(figsize=(40, 16))
plt.plot(df_group_city0_dt['date_dt'], df_group_city0_dt['dwell'], marker='o')
plt.grid(color='#95a5a6',linestyle='--', linewidth=1,axis='both',alpha=0.4)
```
